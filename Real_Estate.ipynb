```python
Real_estate=pd.read_csv('Real Estate Asset Management Data_Migrated Data.csv')
```

```python
Real_estate.head(5)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
#another way of encoding
#from sklearn import prepocessing
#le=preprocessing .LabelEncoder
#df['Asset Manager']=le.fit.transform(df['Asset Manager'])

encoded_columns=['Asset Manager','Prop ID','Property Name','Property Type','Average Occupancy %']

for i in encoded_columns:

  Real_estate[i]=le.fit_transform(Real_estate[i])
  print(i)
```

```python
#feature engineering(feature engineering is identifying the target variable and classifying the variables ,encoding and getting te model) and drop some of the irrelevant columns
Real_estate=Real_estate.drop(columns='Prop ID')
```

```python
Real_estate=Real_estate.drop(columns='Property Name')
```

```python
x=Real_estate.drop(columns='Average Occupancy %',axis=1)
y=Real_estate['Average Occupancy %']
```

```python
# splitting the dataset to training and testing
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
clfi_gini=DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=0)
clfi_gini.fit(x_train,y_train)
```

```python
y_pred=clfi_gini.predict(x_test)
```

```python
#check for accuracy
from sklearn.metrics import accuracy_score
print('model accuracy score :{0:0.4f}'.format(accuracy_score(y_test,y_pred)))
```
